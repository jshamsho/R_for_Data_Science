```r
library(tidyverse)
```

```r
# Visualize distribution of categorical variable
ggplot(data = diamonds) +
  geom_bar(mapping = aes(x = cut))
```

```r
# Numerical distribution using dyplyr
diamonds %>%
    count(cut)
```

```r
# Histogram of diamond carat
ggplot(data = diamonds) +
  geom_histogram(mapping = aes(x = carat), binwidth = 0.5)
```

```r
# Numerically
diamonds %>% 
  count(cut_width(carat, 0.5))
```

```r
# On a finer scale
diamonds %>%
    filter(carat < 3) %>%
    ggplot(mapping = aes(x = carat)) + 
    geom_histogram(binwidth = 0.1)
```

```r
# Display multiple histograms using geom_freqpoly
smaller <- diamonds %>%
    filter(carat < 3)
ggplot(data = smaller, mapping = aes(x = carat, colour = cut)) +
  geom_freqpoly(binwidth = 0.1)
```

```r
# Why are there more diamonds slightly to right of each peak?
ggplot(data = smaller, mapping = aes(x = carat)) +
  geom_histogram(binwidth = 0.01)

# How are observations within cluster similar to each other?
# How are the observations in separate clusters different from each other?
```
