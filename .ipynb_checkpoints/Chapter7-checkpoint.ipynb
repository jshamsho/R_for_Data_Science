```r
# Exercises
# Explore the distribution of each of the x, y, and z variables in diamonds.
# What do you learn? Think about a diamond and how you might decide which
# dimension is the length, width, and depth.
par(mfrow = c(1,3))
options(repr.plot.width=16, repr.plot.height=4)
x_plot <- diamonds %>%
    ggplot() +
    geom_histogram(mapping = aes(x = x))
y_plot <- diamonds %>%
    ggplot() +
    geom_histogram(mapping = aes(x = y))
z_plot <- diamonds %>%
    ggplot() +
    geom_histogram(mapping = aes(x = z))
grid.arrange(x_plot, y_plot, z_plot, nrow = 1)
```

```r
# Explore the distribution of price. Do you discover anything unusual or surprising?
# (Hint: Carefully think about the binwidth and make sure you try a wide range of values.)
diamonds %>%
    ggplot(mapping = aes(x = price)) +
    geom_histogram(binwidth = 10) + 
    coord_cartesian(xlim = c(1000, 2000))
# There's few diamonds priced around $1500
```

```r
# How many diamonds are 0.99 carat?
# How many are 1 carat? What do you think is the cause of the difference?
diamonds %>% 
    filter(carat == 0.99) %>%
    summarize(n())

diamonds %>%
    summarize(carat.99 = sum(carat == 0.99))

diamonds %>% 
    filter(carat == 1.00) %>%
    summarize(n())

diamonds %>%
    summarize(carat1.00 = sum(carat == 1.00))
```

```r
# Replace error values with NA codes
diamonds2 <- diamonds %>% 
  mutate(y = ifelse(y < 3 | y > 20, NA, y))
```

```r
# ggplot2 gives warning when plot omits missing values
options(repr.plot.width=4, repr.plot.height=4)
ggplot(data = diamonds2, mapping = aes(x = x, y = y)) + 
  geom_point()
```

```r
# Remove warning
ggplot(data = diamonds2, mapping = aes(x = x, y = y)) + 
  geom_point(na.rm = TRUE)
```

```r
# Compare scheduled departure times for cancelled and non-cancelled flights
options(repr.plot.width=6, repr.plot.height=6)
nycflights13::flights %>% 
  mutate(
    cancelled = is.na(dep_time),
    sched_hour = sched_dep_time %/% 100,
    sched_min = sched_dep_time %% 100,
    sched_dep_time = sched_hour + sched_min / 60
  ) %>% 
  ggplot(mapping = aes(sched_dep_time)) + 
    geom_freqpoly(mapping = aes(colour = cancelled), binwidth = 1/4)
```

```r
# What happens to missing values in a histogram?
# What happens to missing values in a bar chart? Why is there a difference?
options(repr.plot.width=12, repr.plot.height=6)
p1 <- diamonds2 %>% ggplot(mapping = aes(x = y)) +
          geom_histogram(na.rm = T)
flights2 <- nycflights13::flights %>% 
  mutate(
    cancelled = is.na(dep_time),
    sched_hour = sched_dep_time %/% 100,
    sched_min = sched_dep_time %% 100,
    sched_dep_time = sched_hour + sched_min / 60
  )
flights2$cancelled[1] <- NA
p2 <- ggplot(flights2, mapping = aes(x = cancelled)) +
          geom_bar()
grid.arrange(p1, p2, nrow = 1)
```

```r
# What does na.rm = TRUE do in mean() and sum()?
diamonds2 %>% 
    summarize(mean(y))
diamonds2 %>%
    summarize(mean(y, na.rm = T))
```

```r
# Covariation
# How does price vary within diamond cut?
ggplot(data = diamonds, mapping = aes(x = price)) + 
  geom_freqpoly(mapping = aes(colour = cut), binwidth = 500)
```

```r
# Display density instead of raw count
ggplot(data = diamonds, mapping = aes(x = price, y = ..density..)) + 
  geom_freqpoly(mapping = aes(colour = cut), binwidth = 500)
```

```r
# Boxplot of price by cut
ggplot(data = diamonds, mapping = aes(x = reorder(cut, price, FUN = median), y = price)) +
  geom_boxplot()
```

```r
# Highway mpg by class
ggplot(data = mpg) +
  geom_boxplot(mapping = aes(x = reorder(class, hwy, FUN = median), y = hwy))
```
